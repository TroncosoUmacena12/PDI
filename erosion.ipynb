UTILIZAR SKIMAGE CON EROSION DILATACION APERTURA Y CERRADURA 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology

# Leer la imagen en escala de grises con OpenCV
imagen = cv2.imread('2.png', cv2.IMREAD_GRAYSCALE)
_, imagen_bin = cv2.threshold(imagen, 127, 255, cv2.THRESH_BINARY)

imagen_bin_bool = imagen_bin == 255  

selem = morphology.disk(1)
erosion_img = morphology.erosion(imagen_bin_bool, selem)
dilatacion_img = morphology.dilation(imagen_bin_bool, selem)
apertura_img = morphology.opening(imagen_bin_bool, selem)
cerradura_img = morphology.closing(imagen_bin_bool, selem)

# Mostrar los resultados
plt.figure(figsize=(10, 5))

plt.subplot(1, 5, 1)
plt.imshow(imagen_bin_bool, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(erosion_img, cmap='gray')
plt.title('Erosión')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(dilatacion_img, cmap='gray')
plt.title('Dilatación')
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(apertura_img, cmap='gray')
plt.title('Apertura')
plt.axis('off')

plt.subplot(1, 5, 5)
plt.imshow(cerradura_img, cmap='gray')
plt.title('Cerradura')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.data import checkerboard

#Imagen con pixeles
imagen = np.zeros((300, 300), dtype=np.uint8)
#
imagen = cv2.circle(imagen, (200, 100), 40, 255, -1)
imagen = cv2.rectangle(imagen, (50, 50), (100, 100), 255, -1)
imagen = cv2.line(imagen, (250, 50), (280, 200), 255, 10)

check = checkerboard()>8.5
check = check.astype(np.uint8)*255

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('Figuras Geometricas')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(check, cmap='gray')
plt.title('checkerboard')
plt.axis('off')


plt.show()

kernel_ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (9, 9))
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (9, 9))
kernel_rect = cv2.getStructuringElement(cv2.MORPH_RECT, (9, 9))

#Aplicando
erode_ellipse = cv2.erode(imagen, kernel_ellipse,iterations=1)
dilate_ellipse = cv2.dilate(imagen, kernel_ellipse, iterations=1)

erode_croos = cv2.erode(imagen, kernel_cross,iterations=1)
dilate_croos = cv2.dilate(imagen, kernel_cross, iterations=1)

erode_rect = cv2.erode(imagen, kernel_rect,iterations=1)
dilate_rect = cv2.dilate(imagen, kernel_rect, iterations=1)

erode_check = cv2.erode(check, kernel_ellipse,iterations=1)
dilate_check = cv2.dilate(check, kernel_ellipse, iterations=1)

erode_check = cv2.erode(check, kernel_rect,iterations=1)
dilate_check = cv2.dilate(check, kernel_rect, iterations=1)

erode_check = cv2.erode(check, kernel_cross,iterations=1)
dilate_check = cv2.dilate(check, kernel_cross, iterations=1)


#Creamos imagen apertura y cerradura
np.random.seed(50)
imagen_ruido = imagen.copy()
ruido = np.random.randint(0, 2, imagen.shape, dtype=np.uint8)*255
#Juntar Excluyente
imagen_ruido = cv2.bitwise_xor(imagen_ruido, ruido)

apertura = cv2.morphologyEx(imagen_ruido, cv2.MORPH_OPEN, kernel_rect)
cerradura = cv2.morphologyEx(imagen_ruido, cv2.MORPH_CLOSE, kernel_rect)

plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(erode_ellipse, cmap='gray')
plt.title('Erode Ellipse')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(dilate_ellipse, cmap='gray')
plt.title('Dilate Ellipse')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(erode_croos, cmap='gray')
plt.title('Erode Cross')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(dilate_croos, cmap='gray')
plt.title('Erode Cross')
plt.axis('off')

plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(erode_rect, cmap='gray')
plt.title('Erode Rect')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(dilate_rect, cmap='gray')
plt.title('Dilate Rect')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(erode_check, cmap='gray')
plt.title('Erode Check')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(dilate_check, cmap='gray')
plt.title('Dilate Check')
plt.axis('off')

plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(apertura, cmap='gray')
plt.title('Apertura')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cerradura, cmap='gray')
plt.title('Cerradura')
plt.axis('off')


plt.show()

```
