```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def falso_color(imagen_gray,num):
    list_maps = [cv2.COLORMAP_AUTUMN,cv2.COLORMAP_BONE,cv2.COLORMAP_JET,cv2.COLORMAP_WINTER,cv2.COLORMAP_RAINBOW,cv2.COLORMAP_OCEAN,cv2.COLORMAP_SUMMER,cv2.COLORMAP_SPRING,cv2.COLORMAP_COOL,
                 cv2.COLORMAP_HSV,cv2.COLORMAP_PINK,cv2.COLORMAP_HOT,cv2.COLORMAP_PARULA,cv2.COLORMAP_MAGMA,cv2.COLORMAP_INFERNO,cv2.COLORMAP_PLASMA,cv2.COLORMAP_VIRIDIS,cv2.COLORMAP_CIVIDIS,
                 cv2.COLORMAP_TWILIGHT,cv2.COLORMAP_TWILIGHT_SHIFTED,cv2.COLORMAP_TURBO,cv2.COLORMAP_DEEPGREEN]
    
    escala_imagen = cv2.normalize(imagen_gray,None,0,255,cv2.NORM_MINMAX)
    apli_color = cv2.applyColorMap(escala_imagen,list_maps[num])
    return apli_color


imagen = cv2.imread('1.png')
img_gris = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
cont = 0

list_noms = ['COLORMAP_AUTUMN','COLORMAP_BONE','COLORMAP_JET','COLORMAP_WINTER','COLORMAP_RAINBOW','COLORMAP_OCEAN','COLORMAP_SUMMER','COLORMAP_SPRING','COLORMAP_COOL','COLORMAP_HSV','COLORMAP_PINK',
             'COLORMAP_HOT','COLORMAP_PARULA','COLORMAP_MAGMA','COLORMAP_INFERNO','COLORMAP_PLASMA','COLORMAP_VIRIDIS','COLORMAP_CIVIDIS','COLORMAP_TWILIGHT','COLORMAP_TWILIGHT_SHIFTED',
             'COLORMAP_TURBO','COLORMAP_DEEPGREEN']



num_colormaps = len(list_noms)
cols = 5 
rows = (num_colormaps // cols) + (num_colormaps % cols > 0)  

fig, axes = plt.subplots(rows, cols, figsize=(15, 10))
axes = axes.flatten()

for i in range(num_colormaps):
    img_nueva = falso_color(img_gris, i)
    axes[i].imshow(cv2.cvtColor(img_nueva, cv2.COLOR_BGR2RGB))
    axes[i].set_title(list_noms[i], fontsize=8)
    axes[i].axis('off')

for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
