ENFOQUE

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('1.png')

mascara = np.array([[0, -1, 0],
                    [-1, 5, 1],
                    [0, -1, 0]]) 

img_fin = cv2.filter2D(imagen, -1, mascara)

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen original')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB))
plt.title('Imagen con Enfoque')

plt.show()
```

LISTA DE MATRICES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def mostrar_imagen(imagen, mascara, titulo_filtro):
    img_fin = cv2.filter2D(imagen, -1, mascara)

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
    plt.title('Imagen original')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB))
    plt.title(titulo_filtro)
    plt.axis('off')

    plt.show()
def enfoque(imagen):
    mascara = np.array([[0, -1, 0],
                        [-1, 5, 1],
                        [0, -1, 0]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Enfoque')

def desfoque(imagen):
    mascara = np.array([[1, 1, 1],
                        [1, 1, 1],
                        [1, 1, 1]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Desenfoque')

def realce_borde(imagen):
    mascara = np.array([[0, 0, 0],
                        [-1, 1, 0],
                        [0, 0, 0]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Realce de Borde')

def repujado(imagen):
    mascara = np.array([[-2, -1, 0],
                        [-1, 1, 1],
                        [0, 1, 2]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Repujado')

def deteccion_bordes(imagen):
    mascara = np.array([[0, 1, 0],
                        [1, -4, 1],
                        [0, 1, 0]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Detección de Bordes')

def filtro_sobel(imagen):
    mascara = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Filtro Sobel')

def filtro_sharpen(imagen):
    mascara = np.array([[1, -2, 1],
                        [-2, 5, -2],
                        [1, -2, 1]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Filtro Sharpen')

def filtro_norte(imagen):
    mascara = np.array([[1, 1, 1],
                        [1, -2, 1],
                        [-1, -1, -1]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Filtro Norte')


def filtro_este(imagen):
    mascara = np.array([[-1, 1, 1],
                        [-1, -2, 1],
                        [-1, 1, 1]]) 
    mostrar_imagen(imagen, mascara, 'Imagen con Filtro Este')

def filtro_gauss(imagen):
    mascara = np.array([[1, 2, 3, 1, 1],
                        [2, 7, 11, 7, 2],
                        [3, 11, 17, 11, 3],
                        [2, 7, 11, 7, 1],
                        [1, 2, 3, 2, 1]]) / 49
    mostrar_imagen(imagen, mascara, 'Imagen con Filtro Gauss')

imagen = cv2.imread('1.png')

enfoque(imagen)
desfoque(imagen)
realce_borde(imagen)
repujado(imagen)
deteccion_bordes(imagen)
filtro_sobel(imagen)
filtro_sharpen(imagen)
filtro_norte(imagen)
filtro_este(imagen)
filtro_gauss(imagen)
```
