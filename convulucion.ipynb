```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convulsion_media_horizontal(imagen, tamañoKernel):
    kernel = np.ones([1, tamañoKernel], np.float32)/tamañoKernel
    img_fin = cv2.filter2D(imagen, -1, kernel)
    return img_fin 

imagen = cv2.imread('4.png')
img_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

img_rest_0 = convulsion_media_horizontal(img_rgb[:,:,0], 150)
img_rest_1 = convulsion_media_horizontal(img_rgb[:,:,1], 150)
img_rest_3 = convulsion_media_horizontal(img_rgb[:,:,2], 150)

imagen_color = np.stack([img_rest_0, img_rest_1, img_rest_3], axis=2)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Suavizado 5x5')
plt.imshow(imagen_color)

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Función para convolución con un kernel diagonal
def convulsion_media_diagonal(imagen, tamañoKernel):
    # Crear un kernel diagonal (tamañoKernel x tamañoKernel)
    kernel = np.zeros((tamañoKernel, tamañoKernel), np.float32)
    np.fill_diagonal(kernel, 1.0 / tamañoKernel)  # Llenamos la diagonal con 1/tamañoKernel
    img_fin = cv2.filter2D(imagen, -1, kernel)
    return img_fin 

# Leer la imagen y convertirla a escala de grises
imagen = cv2.imread('5.png')
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplicar la convolución diagonal con un kernel de tamaño 25x25
img_rest = convulsion_media_diagonal(gris, 25)

# Mostrar las imágenes
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Suavizado Diagonal 25x25')
plt.imshow(img_rest, cmap='gray')

plt.show()EJEEEJEEjEEEJSDHFSEJasdasdasdasdasdasd
```
